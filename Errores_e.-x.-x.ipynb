```python
import matplotlib.pyplot as pl
import math

cof=0.567143
rb=[]
tb=[]
rf=[]
tf=[]
rp=[]
tp=[]
rn=[]
tn=[]
rs=[]
ts=[]
ervb=[]
ervf=[]
ervp=[]
ervn=[]
ervs=[]

def biseccion():
    print ("Metodo por Biseccion")
    ci=0
    cs=5
    cm=0
    res=0
    c=0
    while (res==0):
        print ('lim inf',ci,' lim sup',cs)
        fci=(math.exp(-ci))-ci
        fcs=(math.exp(-cs))-cs
        if ((fci*fcs)==0):
            c=cs
            print ("resultado",c)
            break
        else:
            #print ('entro')
            cm= (ci+cs)/2
            if (((math.exp(-cm))-cm)>0):
                ci=cm
            else:
                cs=cm
        errv=cs-cof
        ervb.append(errv)
        
def err_b():
    cont=0
    while (cont<len(ervb)):
        rb.append((ervb[cont]/cof)*100)
        tb.append(cont)
        cont+=1
            
            
def met_falsa_posicion():
    print ("Metodo por Falsa Posicion")
    ci=0
    cs=5
    cm=0
    res=0
    c=0
    while (res==0):
        print ('lim inf',ci,' lim sup',cs)
        fci=(math.exp(-ci))-ci
        fcs=(math.exp(-cs))-cs
        cm=cs-((fcs*(ci-cs))/(fci-fcs))
        if (cm>0):
            cs=cm
        else:
            ci=cm
        if ((fci*fcs)==0):
            c=cs
            print ("resultado",c)
            break
        errv=cs-cof
        ervf.append(errv)
            
def err_f():
    cont=0
    while (cont<len(ervf)):
        rf.append((ervf[cont]/cof)*100)
        tf.append(cont)
        cont+=1
        
def met_punto_fijo():
    print ("Metodo Punto Fijo")
    ci=0
    cf=0
    res=0
    while (res<10):
        if(((math.exp(-ci))-cf)==0):
            print (cf)
            break;
        else:
            cf=math.exp(-ci)
            ci=cf
            print ("Valor",cf)
            res+=1
        errv=cf-cof
        ervp.append(errv)
    
def err_p():
    cont=0
    while (cont<len(ervp)):
        rp.append((ervp[cont]/cof)*100)
        tp.append(cont)
        cont+=1
        
def met_newton_raphson():
    print ("Metodo Newton-Raphson")
    ci=0
    cf=0
    res=0
    while (res<5):
        fci=(ci-(((math.exp(-ci))-ci)/((-1*(math.exp(-ci)))-1)))
        if(fci==0):
            print (cf)
            break;
        else:
            cf=fci
            ci=cf
            print ("Valor",cf)
            res+=1 
        errv=cf-cof
        ervn.append(errv)
            
def err_n():
    cont=0
    while (cont<len(ervn)):
        rn.append((ervn[cont]/cof)*100)
        tn.append(cont)
        cont+=1
        
def met_secante():
    print ("Metodo Secante")
    ca=0
    ci=1
    cf=0
    res=0
    while (res<5):
        fci=( ci-( (((math.exp(-ci))-ci)*(ca-ci)) / ( ((math.exp(-ca))-ca) - ((math.exp(-ci))-ci) ) ) )
        if(fci==0):
            print (cf)
            break;
        else:
            ca=ci
            cf=fci
            ci=cf
            print ("Valor x+1",cf)
            res+=1 
        errv=cf-cof
        ervs.append(errv)
            
def err_s():
    cont=0
    while (cont<len(ervs)):
        rs.append((ervs[cont]/cof)*100)
        ts.append(cont)
        cont+=1

def graf():
    
    pl.ylabel(' ')
    pl.xlabel('Tiempo (s)')
    pl.plot(tb,rb,"-")
    pl.plot(tf,rf,"-")
    pl.plot(tp,rp,"-")
    pl.plot(tn,rn,"-")
    pl.plot(ts,rs,"-")
    pl.grid(True)
    pl.axis([-10,50,0,100])
    pl.show() 

biseccion()
met_falsa_posicion()
met_punto_fijo()
met_newton_raphson()
met_secante()
err_b()
err_f()
err_p()
err_n()
err_s()
graf()
```
