```python
import math
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
%matplotlib inline
vectorA=[10,16.30,23,27.50,31,35.6,39,41.5,42.9,45,46,45.5,46,49,50]
vectorB=[8.953,16.405,22.607,27.769,32.065,35.641, 38.617,41.095, 43.156,44.872,46.301, 47.490, 48.479, 49.303, 49.988]
vectorC=[11.240,18.570, 23.729, 27.556, 30.509, 32.855, 34.766, 36.351, 37.687, 38.829, 39.816, 40.678, 41.437, 42.110, 42.712]
n=len(vectorA)
t,z,t2,ac=[],[],[],[]
def suma (a):
    s=0
    for i in range(n):
        s+=a[i]
    return s
def multiplicacion(a,b):
    r=0
    for j in range (n):
        r+=a[j]*b[j]
        t.append(j)
    return r

num=(n*(multiplicacion(vectorA,vectorB))-(suma(vectorA)*suma(vectorB)))
den=(math.sqrt(n*multiplicacion(vectorA, vectorA)-suma(vectorA)**2)*math.sqrt(n*multiplicacion(vectorB, vectorB)-suma(vectorB)**2))

res=(num)/(den)

num1=(n*(multiplicacion(vectorA,vectorC))-(suma(vectorA)*suma(vectorC)))
den1=(math.sqrt(n*multiplicacion(vectorA, vectorA)-suma(vectorA)**2)*math.sqrt(n*multiplicacion(vectorC, vectorC)-suma(vectorC)**2))

res1=(num1)/(den1)

a1=(n*(multiplicacion(vectorA,vectorB))-(suma(vectorA))*suma(vectorB))/((n*multiplicacion(vectorA,vectorA))-(suma(vectorA)**2))
a0=((suma(vectorB))/n)-a1*((suma(vectorA))/n)

a1b=(n*(multiplicacion(vectorA,vectorC))-(suma(vectorA))*suma(vectorC))/((n*multiplicacion(vectorA,vectorA))-(suma(vectorA)**2))
a0b=((suma(vectorC))/n)-a1b*((suma(vectorA))/n)
for k in range (50):
    y=a0+a1*k
    y2=a0b+a1b*k
    z.append(y)
    ac.append(y2)
    t2.append(k)
print ("coeficiente de correlaci贸n para a,b ",res)
print ("coeficiente de correlaci贸n para a,c ",res1)

plt.plot(vectorA,vectorB, 'ro', label='v medida / v calculada b')
plt.plot(t2,z, label='recta y=a0+a1*x (a y b)')
plt.grid()
plt.xlabel('v medida')
plt.ylabel('y')
plt.legend(loc='upper left', numpoints=1)
plt.show()

plt1.plot(vectorA, vectorC, 'go', label='v medida / v calcula c')
plt1.plot(t2,ac, label='recta Y=a0+a1*x (a y c)')
plt1.grid()
plt1.xlabel('v medida')
plt1.ylabel('y')
plt1.legend(loc='upper left', numpoints=1)
plt1.show()
```
