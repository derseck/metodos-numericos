```python
import math
import matplotlib.pyplot as pl

f=[]
f1=[]
x1=[]
x2=[]
def met_grafico():
    x=0
    for a in range(400):
        f.append((18-2*x)/3)
        f1.append((2*x-2))
        x1.append(x)
        x2.append(x)
        x+=0.1

def met_ecuaciones():
    x=0
    for a in range(200):
        xr1=((-1*3/2)*x)+(18/2)
        xr2=(-1*(-1/2)*x)+(2/2)
        x+=0.5
        print("rta ec1:",xr1)
        print("rta ec2:",xr2)

met_ecuaciones()             
met_grafico()
pl.xlabel("")
pl.ylabel("")
pl.plot(x1,f)
pl.plot(x2,f1)
pl.axis([0,9,0,6])
pl.grid(True)
pl.show()
```
