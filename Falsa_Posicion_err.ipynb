```python
import matplotlib.pyplot as pl
import math

rb=[]
tb=[]
rf=[]
tf=[]
ervb=[]
ervf=[]
cof=14.780203

def biseccion():
    g=9.8
    m=68.1
    t=10
    v=40
    ci=4
    cs=15
    cm=0
    res=0
    c=0
    while (res==0):
        print ('lim inf',ci,' lim sup',cs)
        fci=((g*m/ci)*(1-math.exp(-(ci/m)*t)))-v
        fcs=((g*m/cs)*(1-math.exp(-(cs/m)*t)))-v
        if ((fci*fcs)==0):
            c=cs
            print (c)
            break
        else:
            #print ('entro')
            cm= (ci+cs)/2
            if ((((g*m/cm)*(1-math.exp(-(cm/m)*t)))-v)>0):
                ci=cm
            else:
                cs=cm
        errv=cs-cof
        ervb.append(errv)

def err_b():
    cont=0
    while (cont<len(ervb)):
        rb.append((ervb[cont]/cof)*100)
        tb.append(cont)
        cont+=1
                
def met_falsa_posicion():
    g=9.8
    m=68.1
    t=10
    v=40
    ci=12
    cs=16
    cm=0
    res=0
    c=0
    while (res==0):
        print ('lim inf',ci,' lim sup',cs)
        fci=((g*m/ci)*(1-math.exp(-(ci/m)*t)))-v
        fcs=((g*m/cs)*(1-math.exp(-(cs/m)*t)))-v
        cm=cs-((fcs*(ci-cs))/(fci-fcs))
        if (cm>0):
            cs=cm
        else:
            ci=cm
        if ((fci*fcs)==0):
            c=cs
            print ("resultado",c)
            break
        errv=cs-cof
        ervf.append(errv)

def err_f():
    cont=0
    while (cont<len(ervf)):
        rf.append((ervf[cont]/cof)*100)
        tf.append(cont)
        cont+=1            
            
def graf():
    
    pl.ylabel(' ')
    pl.xlabel('Tiempo (s)')
    pl.plot(tb,rb,"-")
    pl.plot(tf,rf,"-")
    pl.grid(True)
    pl.axis([0,10,-10,10])
    pl.show()
    
biseccion()
met_falsa_posicion()
err_b()
err_f()
graf()
```
